---

*2. Creating Additional Columns*

---



```python
# Convert 'rooms' to numeric, forcing errors to NaN (in case of any invalid data)
df['rooms'] = pd.to_numeric(df['rooms'], errors='coerce')

# Create a new column 'property_size_category' based on the 'rooms' column
def categorize_property_size(rooms):
    if pd.isna(rooms):
        return 'Unknown'  # Handle NaN values by categorizing as 'Unknown'
    elif rooms <= 1:
        return 'Small'
    elif rooms <= 3:
        return 'Medium'
    else:
        return 'Large'

df['property_size_category'] = df['rooms'].apply(categorize_property_size)

# Verify the new column
print(df[['rooms', 'property_size_category']].head())
```
